```python
for cols in df.select_dtypes(include='number').columns:
    fig = go.Figure()
    fig.add_trace(go.Box(y=df[cols].dropna(), name=cols, boxpoints='all', jitter=0.3, pointpos=-1.8))
    fig.update_layout(
        title=f'Distribution of {cols}',
        xaxis_title='',
        yaxis_title='Frequency',
        width=800,
        height=400
    )
    fig.show()
```
